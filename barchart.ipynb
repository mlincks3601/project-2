```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
```

```python
# Load in csv
tour_df = pd.read_csv("tour.csv")
tour_df.head()
```

```python
new_tour_df = tour_df.set_index("year")
new_tour_df
```

```python
# Split up our data into groups based upon 'band'
grouped_tour_df = new_tour_df.groupby('year')

# Find out how many of each date took bike trips
value_count_df = grouped_tour_df['band'].value_counts()

# Drop the 'stoptime' row that is contained within our group
value_count_df = value_count_df.drop(columns =["id", "city_id", "city", "state", "country", "lat", "long"])

# Chart our data, give it a title, and label the axes
bar_chart_df = value_count_df.plot(kind = "barh", title = "Concerts by Year")
bar_chart_df.set_xlabel("Number of Concerts")
bar_chart_df.set_xticks([0,5,10,15,20,25,30,35,40,45,50,55,60,65,70])
bar_chart_df.set_ylabel("bands")

plt.show()
plt.tight_layout()
```
